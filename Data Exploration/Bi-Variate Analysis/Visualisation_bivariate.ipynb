```python
# Functin to print the total percentage of the missing values.

def missing_percentage(df):
    """This function takes a DataFrame(df) as input and returns two columns, total 
    missing values and total missing values percentage"""
    
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2)
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

%matplotlib inline
```

```python
def combination_plots(df,cols):
    
    num = int(input('Enter 1.Boxplot /n 2.Violin Plot /n 3. Swarm Plot '))
    
    if(num==1):
        ## Plotting Boxplot for the Data
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        fig=sns.boxplot(x= cols[0],y=cols[1],data=df,linewidth=2.5,orient='v',dodge=False)
        
    elif(num==2):
        ### Violin Plot for the data
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        fig=sns.violinplot(x= cols[0],y=cols[1],data=df)
        
    elif(num==3):
        ## Swarm plot for the data :
        sns.set(style="whitegrid")
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        ax = sns.swarmplot(x= cols[0],y=cols[1], data=df,color="white", edgecolor="black") 
        
    else:
        print('Enter Currect input for Plotting')
```

```python
# taking multiple inputs at a time  

x = [x for x in input("Enter multiple Columns: ").split()] 
combination_plots(data,x)
```

```python
def correlation_plots(df,cols):
    
    num = int(input('Enter 1.Scatterplot /n 2.Joint Plot /n 3. Residual Plot '))
    
    if(num==1):
        ## Plotting Scatter for the Data
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        fig = sns.scatterplot(x= cols[0],y=cols[1],data=df)
    elif(num==2):
        ### Joint Plot for the data
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        fig = sns.jointplot(x= cols[0],y=cols[1],data=df, kind="reg")
        
    elif(num==3):
        ## Residual plot for the data :
        sns.set(style="whitegrid")
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        fig=sns.residplot(x= cols[0],y=cols[1],data=df,lowess=True) 
        
    else:
        print('Enter Currect input for Plotting')
```

```python
# taking multiple Numerical column inputs at a time  

x = [x for x in input("Enter multiple Columns: ").split()] 
correlation_plots(data,x)
```

```python
def Association_plots(df,cols):
    
    num = int(input('Enter 1.Heat Map /n 2.Stacked Column Chart /n '))
    
    cor = df.pivot(cols[0],cols[1],cols[2])
    colors = ["#006D2C", "#31A354","#74C476"]
    if(num==1):
        
        ## Heat Map for the Data
        fig=plt.gcf()
        fig.set_size_inches(10,7)
        #cor = df.pivot(cols[0],cols[1],cols[2])
        fig = sns.heatmap(cor, vmax=.3, square=True,  cmap="YlGnBu" ) 
        
    elif(num==2):
        
        ## Stacked Column Chart
        fig = cor.plot.bar(stacked=True, color=colors, figsize=(10,7))
        
    else:
        
        print('Enter Correct number')
```
